```python
from iterstrat.ml_stratifiers import MultilabelStratifiedKFold
```

```python
df = train_targets_scored
df.loc[:, "kfold"] = -1

df = df.sample(frac=1).reset_index(drop=True)
y = df.drop("sig_id", axis=1).values

msk = MultilabelStratifiedKFold(n_splits=5)

for fold, (trn_idx, val_idx) in enumerate(msk.split(X=df, y=y)):
    df.loc[val_idx, "kfold"] = fold

df.to_csv("../input/train_targets_folds.csv", index=False)
```

```python
train_targets_scored = pd.read_csv("../input/train_targets_folds.csv")
train_targets_scored.head()
```

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import numpy as np
```

```python
class MoaDataset(Dataset):
    def __init__(self, features, targets):
        self.features = features
        self.targets = targets
        
    def __len__(self):
        return self.features.shape[0]
        
    def __getitem__(self, index):
        return {
            "x": torch.tensor(self.features[index, :], dtype=torch.float), 
            "y": torch.tensor(self.targets[index, :], dtype=torch.float),
        }
    
class TestDataset(Dataset):
    def __init__(self, features):
        self.features = features
        
    def __len__(self):
        return self.features.shape[0]
        
    def __getitem__(self, index):
        return {
            "x": torch.tensor(self.features[index, :], dtype=torch.float), 
        }
```

```python
class Model(nn.Module):
    def __init__(self, num_features, num_targets, hidden_size):
        super(Model, self).__init__()
        self.fc1 = nn.Linear(num_features, hidden_size)
        self.fc2 = nn.Linear(hidden_size, hidden_size)
        self.fc3 = nn.Linear(hidden_size, num_targets)
        self.bn1 = nn.BatchNorm1d(num_features)
        self.bn2 = nn.BatchNorm1d(hidden_size)
        self.drop = nn.Dropout(0.2)
        self.relu = nn.ReLU()
        
    def forward(self, x):
        x = self.fc1(self.bn1(x))
        x = self.relu(x)
        x = self.fc2(self.drop(self.bn2(x)))
        x = self.relu(x)
        x = self.fc3(self.bn2(x))
        
        return x
```

```python
def train(epoch, model, device, loader, optimizer, loss_fn):
    model.train()
    final_loss = 0
    for data in loader:
        optimizer.zero_grad()
        inputs = data["x"].to(device)
        targets = data["y"].to(device)
        
        outputs = model(inputs)
        loss = loss_fn(outputs, targets)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        final_loss += loss.item()
        
    final_loss = final_loss / len(loader)
    
    return final_loss
```

```python
def validate(model, device, loader, loss_fn):
    model.eval()
    final_loss = 0
    valid_preds = []
    for data in loader:
        inputs = data["x"].to(device)
        targets = data["y"].to(device)
        
        outputs = model(inputs)
        loss = loss_fn(outputs, targets)
        
        final_loss += loss.item()
        valid_preds.append(outputs.sigmoid().detach().cpu().numpy())
    
    final_loss = final_loss / len(loader)
    valid_preds = np.concatenate(valid_preds)
    
    return final_loss, valid_preds
```

```python
def inference(model, device, loader):
    model.eval()
    preds = []
    for data in loader:
        inputs = data["x"].to(device)
        with torch.no_grad():
            outputs = model(inputs)
        
        preds.append(outputs.sigmoid().detach().cpu().numpy())
    
    preds = np.concatenate(preds)
    
    return preds
```

```python
def run(fold):
    DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    BATCH_SIZE = 128
    EPOCHS = 5
    LEARNING_RATE = 1e-3
    
    BEST_LOSS = np.inf
    EARLY_STOPPING_ITER = 10
    EARLY_STOPPING_COUNTER = 0
    
    df = train_features
    targets = train_targets_scored
    
    feature_columns = df.drop("sig_id", axis=1).columns
    target_columns = targets.drop(["sig_id", "kfold"], axis=1).columns
    
    df = pd.merge(
        df, 
        targets,
        on="sig_id",
        how="left",
    )

    train_df = df[df.kfold != fold].reset_index(drop=True)
    valid_df = df[df.kfold == fold].reset_index(drop=True)
    
    xtrain = train_df[feature_columns].to_numpy()
    ytrain = train_df[target_columns].to_numpy()
    xvalid = valid_df[feature_columns].to_numpy()
    yvalid = valid_df[target_columns].to_numpy()
    
    train_set = MoaDataset(xtrain, ytrain)
    val_set = MoaDataset(xvalid, yvalid)
    
    train_loader = DataLoader(
        train_set,
        batch_size=BATCH_SIZE,
        shuffle=True,
    )
    valid_loader = DataLoader(
        val_set,
        batch_size=BATCH_SIZE,
        shuffle=False,
    )
    
    model = Model(
        num_features=xtrain.shape[1], 
        num_targets=ytrain.shape[1], 
        hidden_size=1024,
    )
    model.to(DEVICE)
    
    optimizer = torch.optim.Adam(
        params=model.parameters(),
        lr=LEARNING_RATE,
    )
    
    loss_fn = nn.BCEWithLogitsLoss()
    
    print("Start training...")
    for epoch in range(EPOCHS):
        train_loss = train(
            epoch, 
            model, 
            DEVICE, 
            train_loader, 
            optimizer, 
            loss_fn,
        )
        
        valid_loss, valid_preds = validate(
            model, 
            DEVICE, 
            valid_loader, 
            loss_fn,
        )
        
        print(f"Fold: {fold}, Epoch: {epoch}, train_Loss: {train_loss}, valid_Loss: {valid_loss}")
        
        if valid_loss < BEST_LOSS:
            BEST_LOSS = valid_loss
            torch.save(model.state_dict(), f"model_{fold}.bin")
        else:
            EARLY_STOPPING_COUNTER += 1
            
        if EARLY_STOPPING_COUNTER > EARLY_STOPPING_ITER:
            break
        
    print("Start inference...")
    
    test = test_features
    
    xtest = test_features[feature_columns].to_numpy()
    test_set = TestDataset(xtest)
    
    test_loader = DataLoader(
        test_set,
        batch_size=BATCH_SIZE,
        shuffle=False,
    )
    
    model = Model(
        num_features=xtrain.shape[1], 
        num_targets=ytrain.shape[1], 
        hidden_size=1024,
    )
    model.load_state_dict(torch.load(f"model_{fold}.bin"))
    model.to(DEVICE)
    
    pred = inference(
        model, 
        DEVICE, 
        test_loader,
    )
    
    pred_df = pd.DataFrame()
    pred_df["sig_id"] = test_features["sig_id"].values
    targets_columns = [col for col in train_targets_scored.columns if col not in ("sig_id", "kfold")]
    for col in targets_columns:
        pred_df[col] = 0   
    pred_df.loc[:, 1:] = pred
    
    pred_df.to_csv(f"../input/pred_{fold}.csv", index=False)
    print("output csv file !!")
    
if __name__ == "__main__":
    for j in range(1):
        run(j)
```
